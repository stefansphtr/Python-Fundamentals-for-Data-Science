3. In addition to organizing the hurricanes in a dictionary with names as the key, you want to be able to organize the hurricanes by year.

   Write a function that converts the current dictionary of hurricanes to a new dictionary, where the keys are years and the values are lists containing a dictionary for each hurricane that occurred in that year.

   For example, the key `1932` would yield the value: `[{'Name': 'Bahamas', 'Month': 'September', 'Year': 1932, 'Max Sustained Wind': 160, 'Areas Affected': ['The Bahamas', 'Northeastern United States'], 'Damage': 'Damage not recorded', 'Deaths': 16}, {'Name': 'Cuba II', 'Month': 'November', 'Year': 1932, 'Max Sustained Wind': 175, 'Areas Affected': ['Lesser Antilles', 'Jamaica', 'Cayman Islands', 'Cuba', 'The Bahamas', 'Bermuda'], 'Damage': 40000000.0, 'Deaths': 3103}]`.

   Test your function on your hurricane dictionary.


````python
# 3
# Organizing by Year


def organize_by_year(
    names, months, years, max_sustained_winds, areas_affected, updated_damages, deaths
):
    """
    Organizes hurricane data by year.

    This function takes in lists of hurricane attributes and organizes them into a dictionary where each key is a year and the value is another dictionary containing the hurricane attributes for that year.

    Parameters:
    names (list): A list of hurricane names.
    months (list): A list of the months in which the hurricanes occurred.
    years (list): A list of the years in which the hurricanes occurred.
    max_sustained_winds (list): A list of the maximum sustained winds during the hurricanes.
    areas_affected (list): A list of the areas affected by the hurricanes.
    updated_damages (list): A list of the updated damages caused by the hurricanes.
    deaths (list): A list of the number of deaths caused by the hurricanes.

    Returns:
    dict: A dictionary where each key is a year and the value is another dictionary containing the hurricane attributes for that year.
    """
    year_hurricane = {}

    for i, year in enumerate(years):
        year_hurricane[year] = {
            "Name": names[i],
            "Month": months[i],
            "Year": year,
            "Max Sustained Wind": max_sustained_winds[i],
            "Areas Affected": areas_affected[i],
            "Damage": updated_damages[i],
            "Deaths": deaths[i],
        }

    return year_hurricane


# Create a new dictionary of hurricanes with year and key
hurricanes_by_year = organize_by_year(
    names, months, years, max_sustained_winds, areas_affected, updated_damages, deaths
)

# Print the result
print(hurricanes_by_year)
````

4. You believe that knowing how often each of the areas of the Atlantic are affected by these strong hurricanes is important for making preparations for future hurricanes.

   Write a function that counts how often each area is listed as an affected area of a hurricane. Store and return the results in a dictionary where the keys are the affected areas and the values are counts of how many times the areas were affected.

   Test your function on your hurricane dictionary.


````python
def count_affected_areas(hurricanes):
    """
    Count the number of times each area is affected by hurricanes.

    Parameters:
    hurricanes (dict): A dictionary where each key is the name of a hurricane and the value is another dictionary
                       containing details about the hurricane. One of the keys in this inner dictionary is "Areas Affected",
                       which is a list of areas affected by that hurricane.

    Returns:
    dict: A dictionary where each key is an area and the value is the number of times that area was affected by a hurricane.
    """
    # Initialize an empty dictionary to store the count of affected areas
    affected_areas_count = {}

    # Iterate over each hurricane in the input dictionary
    for cane in hurricanes:
        # Iterate over each area in the "Areas Affected" list of the current hurricane
        for area in hurricanes[cane]["Areas Affected"]:
            # If the area is not already in the affected_areas_count dictionary, add it with a count of 1
            if area not in affected_areas_count:
                affected_areas_count[area] = 1
            # If the area is already in the affected_areas_count dictionary, increment its count by 1
            else:
                affected_areas_count[area] += 1

    # Return the dictionary of affected areas and their counts
    return affected_areas_count


# Create a dictionary to store the number of hurricanes each area is involved in
affected_areas_count = count_affected_areas(hurricanes)

# Print the dictionary of affected areas and their counts
print(affected_areas_count)
````

5. Write a function that finds the area affected by the most hurricanes, and how often it was hit.

   Test your function on your affected area dictionary.


````python
def most_affected_area(affected_areas_count):
    """
    Find the most affected area and the number of hurricanes it was involved in.


    Parameters:
    affected_areas_count (dict): A dictionary where each key is an area and the value is the number of times
                                 that area was affected by a hurricane.

    Returns:
    tuple: A tuple containing the most affected area and the number of hurricanes it was involved in.
    """
    # Initialize the most affected area and its count
    max_area = "Central America"
    max_area_count = 0

    # Iterate over each area in the input dictionary
    for area in affected_areas_count:
        # If the current area's count is greater than the max_area_count, update max_area and max_area_count
        if affected_areas_count[area] > max_area_count:
            max_area = area
            max_area_count = affected_areas_count[area]

    # Return the most affected area and its count
    return max_area, max_area_count


# Call the function with the affected_areas_count dictionary to find the most frequently affected area
# and the number of hurricanes it was involved in
max_area, max_area_count = most_affected_area(affected_areas_count)

# Print the most affected area and its count
print(max_area, max_area_count)
````

````python
# 7 ==> Solution 2
# Rating Hurricanes by Mortality
def categorize_by_mortality(hurricanes):
    """
    Categorize hurricanes by mortality and return a dictionary.

    Parameters:
    hurricanes (dict): A dictionary where each key is a hurricane name and the value is another dictionary
                       with various properties of the hurricane, including "Deaths".

    Returns:
    dict: A dictionary where each key is a mortality rating and the value is a list of hurricanes that fall into that mortality category.
    """
    mortality_scale = {0: 0, 1: 100, 2: 500, 3: 1000, 4: 10000}
    hurricanes_by_mortality = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []}
    for cane in hurricanes:
        num_deaths = hurricanes[cane]["Deaths"]
        for rating in sorted(mortality_scale.keys(), reverse=True):
            if num_deaths > mortality_scale[rating]:
                hurricanes_by_mortality[rating + 1].append(hurricanes[cane])
                break
    return hurricanes_by_mortality


# categorize hurricanes in new dictionary with mortality severity as key
categorize_hurricanes = categorize_by_mortality(hurricanes)

print(categorize_hurricanes[5])
````
